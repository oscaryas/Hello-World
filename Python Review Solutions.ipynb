```python
#Matrix from example in lecture
M= np.array([[1,0,3],[0,-2,0],[3,0,1]])
print(M)
```

```python
#Compute the eigenvalues of matrix above, and print the values 
w,v = np.linalg.eig(M)
```

```python
#Print Eigenvectors
print(v)
```

## Visualizing Tensors

```python
# Import necessary libraries
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define a helper function to visualize 3D tensor slices
def plot_tensor_slices(tensor, axis=0):
    """
    Visualize slices of a 3D tensor along a specified axis.

    Args:
    tensor (np.ndarray): A 3D tensor to visualize.
    axis (int): Axis along which to slice the tensor (0, 1, or 2).
    """
    # Check if the tensor is 3D
    assert tensor.ndim == 3, "The input tensor must be 3-D"

    # Number of slices depends on the axis chosen
    num_slices = tensor.shape[axis]
    
    fig, axes = plt.subplots(1, num_slices, figsize=(15, 5))
    for i in range(num_slices):
        if axis == 0:
            slice_ = tensor[i, :, :]
        elif axis == 1:
            slice_ = tensor[:, i, :]
        else:
            slice_ = tensor[:, :, i]

        ax = axes[i] if num_slices > 1 else axes
        ax.imshow(slice_, cmap='viridis')
        ax.set_title(f'Slice {i} along axis {axis}')
        ax.axis('off')

    plt.tight_layout()
    plt.show()

# Define a 3D tensor using TensorFlow
tensor_3d_tf = tf.random.uniform(shape=(4, 4, 4), minval=0, maxval=1)

# Convert TensorFlow tensor to NumPy array for visualization
tensor_3d_np = tensor_3d_tf.numpy()

# Visualize the slices of the 3D tensor
plot_tensor_slices(tensor_3d_np, axis=0)  # Slicing along axis 0
```

```python
def plot_3d_scatter(tensor):
    """ This function aims to create a 3D scatter plot of a 3D tensor.
    """
    assert tensor.ndim == 3, "The input tensor must be 3-dimensional."

    # Get the indices and values of the tensor
    x, y, z = np.indices(tensor.shape)
    values = tensor.flatten()

    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')

    # Normalize values for coloring
    colors = plt.cm.viridis(values / values.max())

    # Create a 3D scatter plot
    ax.scatter(x.flatten(), y.flatten(), z.flatten(), c=colors, s=50)

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    plt.title('3D Scatter Plot of Tensor Values')
    plt.show()

# Visualize the 3D tensor using a scatter plot
plot_3d_scatter(tensor_3d_np)
```
